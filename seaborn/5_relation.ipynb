```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip', hue='day', style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {"Lunch":'D',"Dinner":'s'}
sns.scatterplot(data=tips,x='total_bill',y='tip', hue='day', style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip', hue='day', style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
```

## Replot
- Genera una grafica segun le paresca mejor a seaborn 

```python
sns.relplot(data=tips,x='total_bill',y='tip', hue='day', style='time',size='size',col='time')
plt.show()
```

```python
fig, axes = plt.subplots(1,2)
sns.boxenplot(ax=axes[0],data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.scatterplot(ax=axes[1],data=tips,x='total_bill',y='tip', hue='day', style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
```
