```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'bs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.pie(x)
plt.show()
```
