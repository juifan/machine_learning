```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

## Subplots
- Utiliza la programacion orientada a objetos

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

- **nrows** agrega filas a la figura
- **ncols** agrega columnas a la figura

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'m')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'m')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x),'m')
axes[0,2].plot(x, np.tan(x),'b')
axes[0,3].plot(x, np.cos(x)**2)
```

- **tight_layout** Re acomoda la graficas de manera mas estética

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2,ncols=4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x),'m')
ax3.plot(x, np.tan(x),'b')
ax4.plot(x, np.cos(x)**2)
ax5.plot(x, np.cos(x), 'r--')
ax6.plot(x, np.sin(x),'m:')
ax7.plot(x, np.tan(x),'y-.')
ax8.plot(x, np.cos(x)**2, 'm-')
fig.tight_layout()
```
